```python
# load rgb image
my_img1=cv2.imread('new_rgb.png')
cv2_imshow(my_img1)
```

```python
# split blue ch from green and red
b_channel,g_channel,r_channel=cv2.split(my_img1)

# display each channel
cv2_imshow(b_channel)
cv2_imshow(g_channel)
cv2_imshow(r_channel)

cv2_imshow(my_img1)
```

```python
import numpy as np
```

```python
# define a zero matrix
zeros=np.zeros(b_channel.shape,dtype='uint8')

# merge zeros to make BGR image
blueBGR=cv2.merge([b_channel,zeros,zeros])
cv2_imshow(blueBGR)
```

```python
ones=np.ones(g_channel.shape,dtype='uint8')
greenBGR=cv2.merge([ones,g_channel,ones])
cv2_imshow(greenBGR)
```

```python
twos=np.ones(r_channel.shape,dtype='uint8')
redBGR=cv2.merge([twos,twos,r_channel])
cv2_imshow(redBGR)
```

```python
zeros=np.ones(r_channel.shape,dtype='uint8')
redBGR=cv2.merge([zeros,zeros,r_channel])
cv2_imshow(redBGR)
```
