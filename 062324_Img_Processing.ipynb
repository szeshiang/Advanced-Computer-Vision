**1. Crop Image**

```python
plt.imshow(input_image)
```

```python
# define coordinates for cropping
y_start, y_end=25,125 # height range
x_start, x_end=50,200 # width range
cropped_image=input_image[y_start:y_end,x_start:x_end]
cv2_imshow(cropped_image)
```

```python
plt.imshow(cropped_image)
```

```python
# start row is 25% of total height and starting col is 25% of total width
# end row is 75% of total height and end col is 75% of total width

height, width=input_image.shape[:2]

start_row,start_col=int(height*0.25),int(width*0.25)
end_row,end_col=int(height*0.75),int(width*0.75)

cropped_image=input_image[start_row:end_row,start_col:end_col]

cv2_imshow(cropped_image)
```

**2. Resize Image**

```python
width=200
height=300
resized_image=cv2.resize(input_image,(width,height))
cv2_imshow(resized_image)
```

```python
scaling_factor=0.5
resized_image_scale=cv2.resize(input_image,(0,0),fx=scaling_factor,fy=scaling_factor)
cv2_imshow(resized_image_scale)
```

**3. Rotation**

```python
rotated_90_clockwise=cv2.rotate(input_image,cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(rotated_90_clockwise)
```

```python
rotated_180=cv2.rotate(input_image,cv2.ROTATE_180)
cv2_imshow(rotated_180)
```

```python
rotated_90_counterclockwise=cv2.rotate(input_image,cv2.ROTATE_90_COUNTERCLOCKWISE)
cv2_imshow(rotated_90_counterclockwise)

```

3a. Rotate image around its centre

```python
# divide by 2 to rotate the img around its centre. Get to define angle of rotation by defining the axes e.g. width/x and height/y etc.
# called Depth rotation matrix

height,width=input_image.shape[:2]
rotation_matrix=cv2.getRotationMatrix2D((width/2,height/2),90,1) # default rotation is counterclockwise
rotated_image=cv2.warpAffine(input_image,rotation_matrix,(width,height))
cv2_imshow(rotated_image)
```

```python
height,width=input_image.shape[:2]
rotation_matrix=cv2.getRotationMatrix2D((width/2,height/4),180,1)
rotated_image=cv2.warpAffine(input_image,rotation_matrix,(width,height))
cv2_imshow(rotated_image)
```

**Multiply images**

multiplication of images have more control over the brightness compared to addition and scalar

```python
multiply_image=cv2.multiply(input_image,Matrix_one) # the earlier Matrix_one is *100. too bright.
cv2_imshow(multiply_image)

M=np.ones(input_image.shape,dtype='uint8')*2
mul_image=cv2.multiply(input_image,M)
mul_image=np.clip(mul_image,0,255)
cv2_imshow(mul_image)
```

```python
M=np.ones(input_image.shape,dtype=np.float32)*1.5
mul_image=cv2.multiply(input_image.astype(np.float32),M)
mul_image=np.clip(mul_image,0,255).astype(np.uint8)
cv2_imshow(mul_image)
```

```python
M=np.ones(input_image.shape,dtype=np.float32)*2.5
d_image=cv2.divide(input_image.astype(np.float32),M)
d_image=np.clip(d_image,0,255).astype(np.uint8)
cv2_imshow(d_image)
```

**Bit-wise Operator**

```python
M=np.ones(input_image.shape,dtype='uint8')*100
bit_wise_and=cv2.bitwise_and(input_image,M)
cv2_imshow(bit_wise_and)
```

```python
M=np.ones(input_image.shape,dtype='uint8')*100
bit_wise_or=cv2.bitwise_or(input_image,M)
cv2_imshow(bit_wise_or)
```

**Thresholding**

```python
cv2_imshow(input_image)
```

```python
gray_scale=cv2.cvtColor(input_image,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_scale)
```

supposedly values below 127 goes to 0, above 225

```python
ret,thresh1=cv2.threshold(gray_scale,127,255,cv2.THRESH_BINARY)
cv2_imshow(thresh1)
```

```python
ret,thresh3=cv2.threshold(gray_scale,127,255,cv2.THRESH_TRUNC)
cv2_imshow(thresh3)

ret,thresh4=cv2.threshold(gray_scale,127,255,cv2.THRESH_TOZERO)
cv2_imshow(thresh4)

ret,thresh5=cv2.threshold(gray_scale,127,255,cv2.THRESH_TOZERO_INV)
cv2_imshow(thresh5)
```

**Adaptive Thresholding**

```python
thresh=cv2.adaptiveThreshold(gray_scale,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,3,5)
cv2_imshow(thresh)
```

```python
image=cv2.GaussianBlur(gray_scale,(3,3),0)
thresh=cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,3,5)
cv2_imshow(thresh)
```

```python
_,thresh2=cv2.threshold(image,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU) # based on img histogram, separate background vs foreground
cv2_imshow(thresh2)
```
